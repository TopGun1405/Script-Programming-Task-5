  - 요구사항 1. 생성자에 count 값과 step 값을 인자로 받을 수 있다.
    - count: Counter 인스턴스가 지니는 초기 정수 값
    - step: Counter 인스턴스의 count가 증가되는 증분 (default 값: 1)

    > \>>> c = Counter(10)  
    \>>> d = Counter(10, 2)  

  - 요구사항 2. 다음과 같이 Counter의 인스턴스를 출력을 해주는 __str__() 메소드를 Counter 클래스 내에 구현하시오.

  - 요구사항 3. 다음과 같이 step에 주어진 증분만큼 count를 증가시키는 incr() 메소드를 Counter 클래스 내에 구현하시오.

    > \>>> c()  
    \>>> d()
    \>>> print(c)  
    [Count (step: 1)] 12  
    \>>> print(d)  
    [Count (step: 2)] 14
  - 요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.
    > \>>> c = c + 5  
    \>>> d = d - 5  
    \>>> print(c)  
    [Count (step: 1)] 17  
    \>>> print(d)  
    [Count (step: 2)] 9
  - 요구사항 6. 다음과 같은 관계 연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.
    > \>>> print(c) > 10  
    True  
    \>>> print(d) > 10  
    False  
    \>>> print(c) < 10  
    False  
    \>>> print(d) < 10  
    True  
    \>>> print(c) == 17  
    True  
    \>>> print(d) != 9  
    False  

```python
class Counter() :
    # 요구사항 1
    def __init__(self, count, step = 1) :
        self.count = count
        self.step = step
        
    # 요구사항 2
    def __str__(self) :
        return '[Count (step: {})] {}'.format(self.step, self.count)
    
    # 요구사항 3
    def incr(self) :
        self.count += self.step
        
    # 요궇사항 4
    def __call__(self) :
        self.incr()
        
    # 요구사항 5
    def __add__(self, other) :
        if type(other) == Counter :
            return Counter(self.count + other.count, self.step)
        else :
            return Counter(self.count + other, self.step)
        
    def __sub__(self, other) :
        if type(other) == Counter :
            return Counter(self.count - other.count, self.step)
        else :
            return Counter(self.count - other, self.step)
        
    # 요구사항 6
    def __lt__(self, other) :
        if type(other) == Counter :
            return self.count < other.count
        else :
            return self.count < other
        
    def __gt__(self, other) :
        if type(other) == Counter :
            return self.count > other.count
        else :
            return self.count > other
        
    def __eq__(self, other) :
        if type(other) == Counter :
            return self.count == other.count and self.step == other.step
        else :
            return self.count == other
        
c = Counter(10)
d = Counter(10, 2)

print(c); print(d)
print()

c.incr(); d.incr()
print(c); print(d)
print()

c(); d()
print(c); print(d)
print()

c = c + 5; d = d - 5
print(c); print(d)
print()

print(c > 10); print(d > 10)
print(c < 10); print(d < 10)
print(c == 17); print(d != 9)
```

4. 다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str__(), eliminate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.
  - MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

- __init__()은 MySet 객체가 생성될 때 자동으로 작동되는데, 객체를 생성할 때 인자로 받은 리스트를 생성된 객체 자신(self)에 모두 삽입한 후 eliminate_duplicate를 호출하도록 구현되어 있다.
- __str__()은 print()로 MySet를 출력할 때 호출되는 매직 메소드로, 객체가 가지고 있는 list 내용을 집합의 형태로 원소 하나하나를 형식에 맞춰 출력하도록 구현되어 있다.
- eliminate_duplicate()는 집합(Set)의 성질 중 중복된 원소가 없도록 하기 위해 인자로 받은 리스트에서 중복된 원소가 없도록 리스트 내용을 바꾸도록 구현되어 있다.

5. 4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오.
  - 정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.
  - 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.
  > \>>> u = s | t  
  \>>> print(u)  
  MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9}  
  - 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.
  > \>>> u = s & t  
  \>>> print(u)  
  MySet: {2, 3}  
  - 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.
  > \>>> s = MySet([1, 2, 3])  
  \>>> t = MySet([3, 4, 5])  
  \>>> u = s - t  
  \>>> print(u)  
  MySet: {1, 2}  

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
            
    def __or__(self, other) :
        if type(other) == MySet :
            return MySet(self + other)
            
    def __and__(self, other) :
        if type(other) == MySet :
            newList = [e for e in self if e in other]
            return MySet(newList)
        
    def __sub__(self, other) :
        if type(other) == MySet :
            andList = self & other
            subList = [e for e in self if e not in andList]
            return MySet(subList)
        
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    u = s | t
    print(u)
    u = s & t
    print(u)
    s = MySet([1, 2, 3])
    t = MySet([3, 4, 5])
    u = s - t
    print(u)
```

```python
import requests
import string
import itertools
import collections

class SearchEngine :
    def __init__(self, *urls) :
        self.urls = [e for e in urls]
        self.dictList = []
        self.totalDict = dict()
        
    def addUrl(self, url) :
        self.urls.append(url)
        
    def removeUrl(self, url) :
        if url in self.urls : self.urls.remove(url)
        else : return
        
    def listUrls(self) :
        for e in self.urls : print(e)
    
    def wordsFrequency(self) :
        for each in self.urls :
            req = requests.get(each)
            source = req.text
            punct = string.punctuation
            
            htmlTag = cssJvsc = False
            pureStr = ''
            pureStrList = []
            for i in range(len(source)) :
                css = jvsc = ''
                if source[i] == '<' and not htmlTag :
                    pureStr += ' '
                    if i + 8 <= len(source) :
                        for j in range(i + 1, i + 7) : css += source[j]
                        for k in range(i + 1, i + 8) : jvsc += source[k]
                    if css[ : -1] == 'style' or jvsc[ : -1] == 'script' :
                        cssJvsc = True
                    elif css == '/style' or jvsc == '/script' and cssJvsc :
                        cssJvsc = False
                    if not cssJvsc : htmlTag = True
                elif source[i] == '>' and htmlTag :
                    htmlTag = False
                    continue
                if not htmlTag and not cssJvsc :
                    pureStr += source[i]
                    
            pureStrListS = pureStr.split()
            for i in range(len(pureStrListS)) :
                pureStrListS[i] = pureStrListS[i].strip().strip(punct)
                
            for i in range(len(pureStrListS)) :
                for j in punct :
                    while j in pureStrListS[i] :
                        for k in range(len(pureStrListS[i])) :
                            indx = pureStrListS[i].find(j)
                            pureStrListS[i] = pureStrListS[i][ : indx] + ' ' + \
                            pureStrListS[i][indx + 1 : ]
                            break
            pureStrList = ' '.join(pureStrListS).split()
            
            blkCount = pureStrList.count('')
            for i in range(blkCount) :
                pureStrList.remove('')
            
            wordDict = {word : 0 for word in pureStrList}
            for word in wordDict.keys() :
                for val in pureStrList :
                    if word == val : wordDict[word] += 1
                        
            with open('stop_words.txt', 'r', encoding = 'utf-8') as f :
                stopWordList = f.readlines()
                for i in range(len(stopWordList)) :
                    stopWordList[i] = stopWordList[i][ : -1]
                deleteWordList = [keys for word in stopWordList
                                 for keys in wordDict 
                                 if word == keys or word == keys.lower()]
                
                tempSet = {element for element in deleteWordList}
                deleteWordList = [element for element in tempSet]
                for word in deleteWordList :
                    del wordDict[word]
                    
            self.dictList.append(wordDict)

        sumValList = collections.defaultdict(list)
        for i in range(len(self.dictList)) :
            for key, val in itertools.chain(self.dictList[i].items()) :
                sumValList[key].append(val)

        self.totalDict = {key : 0 for key in sumValList.keys()}
        for key, val in sumValList.items() :
            for element in range(len(val)) :
                self.totalDict[key] += val[element]

        #print(len(self.totalDict))
        #print(self.totalDict)
        return self.totalDict
        
    def getWordsFrequency(self) :
        if not self.totalDict :
            self.wordsFrequency()
        print(self.totalDict)
    
    def getMaxFrequencyWord(self) :
        if not self.totalDict :
            self.wordsFrequency()
        sortedDict = [(key, val) for key, val in self.totalDict.items()]
        sortedDict.sort(key = lambda e : e[1], reverse = True)
        maxFrequencyDict = {element[0] : element[1] for element in sortedDict
                           if element[1] == sortedDict[0][1]}
        print(maxFrequencyDict)
    
    def searchUrlByWord(self, word) :
        wordSplitList = word.split()
        with open('stop_words.txt', 'r', encoding = 'utf-8') as f :
            stopWordList = f.readlines()
            for i in range(len(stopWordList)) :
                stopWordList[i] = stopWordList[i][ : -1]
            deleteWordList = [words for sWord in stopWordList
                              for words in wordSplitList if sWord == words or sWord == words.lower()]
            
            tempSet = {element for element in deleteWordList}
            deleteWordList = [element for element in tempSet]
            for words in deleteWordList :
                wordSplitList.remove(words)
                
        similarityList = []
        for i in range(len(self.urls)) :
            for element in wordSplitList :
                if element in self.dictList[i] :
                    similarityList.append((self.urls[i], self.dictList[i][element]))
                elif element.lower() in self.dictList[i] :
                    similarityList.append((self.urls[i], self.dictList[i][element.lower()]))
                elif element.upper() in self.dictList[i] :
                    similarityList.append((self.urls[i], self.dictList[i][element.upper()]))
                elif element.title() in self.dictList[i] :
                    similarityList.append((self.urls[i], self.dictList[i][element.title()]))
                else : similarityList.append((self.urls[i], 0))
                    
        #print(similarityList)
        similarityList.sort(key = lambda e : e[1], reverse = True)
        #print(similarityList)
        print(word)
        for links in similarityList :
            if similarityList[0][1] == 0 :
                print("검색 결과가 없습니다.")
                break
            if links[1] == similarityList[0][1] :
                print(links[0])
            
                
if __name__ == '__main__' :
    w1 = SearchEngine('http://www.cnn.com', 'http://www.times.com', 'https://www.amazon.com')
    w2 = SearchEngine('http://www.cnn.com', 'http://www.times.com')
    w3 = SearchEngine()
    w1.addUrl('https://github.com')
    w3.addUrl('http://stackoverflow.com')
    w1.removeUrl('http://www.cnn.com')
    w2.removeUrl('http://stackoverflow.com')
    print('-' * 80)
    print("w1's url list")
    w1.listUrls()
    print('-' * 80)
    print("w2's url list")
    w2.listUrls()
    print('-' * 80)
    print("w3's url list")
    w3.listUrls()
    print('-' * 80)
#     w1.getWordsFrequency()
#     w2.getWordsFrequency()
#     w3.getWordsFrequency()
    w1.getMaxFrequencyWord()
    print('-' * 80)
    w2.getMaxFrequencyWord()
    print('-' * 80)
    w3.getMaxFrequencyWord()
    print('-' * 80)
    w1.searchUrlByWord("New York")
    print()
    w1.searchUrlByWord("news")
    print()
    w1.searchUrlByWord("github")
    print()
    w1.searchUrlByWord("studio")
    print('-' * 80)
    w2.searchUrlByWord("CNN")
    print()
    w2.searchUrlByWord("times")
    print('-' * 80)
    w3.searchUrlByWord("Stack")
    print()
    w3.searchUrlByWord("Python")
    print('-' * 80)
```

#### 2. 위 1번 문제에서 정의한 SearchEngine 클래스를 상속하여 SearchEngineWithOrderedWebWords 클래스를 정의하고 슈퍼클래스에 정의된 getWordsFrequency() 메소드를 오버라이드하여 단어 출현 빈도를 내림 차순으로 정렬하여 리스트로 출력하시오.
- 리스트 내의 각 원소는 단어와 빈도를 쌍으로 지니는 튜플이다.
- getWordFrequency() 메소드에 reverse 라는 인자를 만들고 true 또는 false를 인자로 받을 수 있도록 한다.
  - reverse 인자의 디폴트 값을 false 이며, 기본적으로 내림차순 정렬한다.
  - reverse 인자에 true를 넣으면 오름차순으로 정렬한다.
  >\>>> w4.SearchEngineWithOrderedWebWords('http://www.times.com', 'https://www.amazon.com', 'https://github.com')</br>  
  \>>> w4.getWordsFrequency()  
  [('site', 12), ('science', 11), ('hello', 8), ('wprld', 2), ('program', 1), ('python', 1)]  
  \>>> w4.getWordsFrequency(reverse=true)  
  [('program', 1), ('python', 1), ('world', 2), ('hello', 8), ('science', 11), ('site', 12)]  

```python
class SearchEngineWithOrderedWebWords(SearchEngine) :
    def __init__(self, *urls) :
        self.urls = [e for e in urls]
        self.dictList = []
        self.totalDict = dict()
        self.sortedList = []

    def getSortedWordsFrequency(self, reverse = False) :
        if not self.totalDict :
            self.wordsFrequency()
        sortedDictList = [(key, val) for key, val in self.totalDict.items()]
        sortedDictList.sort(key = lambda e : e[1], reverse = not reverse)
        self.sortedList = sortedDictList
    
    def getWordsFrequency(self, reverse = False) :
        self.getSortedWordsFrequency(reverse)
        print(self.sortedList)
        
if __name__ == "__main__" :
    w4 = SearchEngineWithOrderedWebWords('http://www.times.com', 
                                         'https://www.amazon.com', 'https://github.com')
    w4.getWordsFrequency()
    w4.getWordsFrequency(reverse = True)
```

#### 3. 다음과 같은 코딩이 가능하도록 SearchEngineWithOrederedWebWords안에 반복자와 관련된 메소드를 추가하시오.

```python
class SearchEngineWithOrderedWebWords(SearchEngine) :
    def __init__(self, *urls) :
        self.urls = [e for e in urls]
        self.dictList = []
        self.totalDict = dict()
        self.sortedList = []
    
    def __iter__(self) :
        self.index = 0
        return self
    
    def __next__(self) :
        if self.index < len(self.sortedList) :
            self.index += 1
            return self.sortedList[self.index - 1]
        else :
            raise StopIteration
    
    def getSortedWordsFrequency(self, reverse = False) :
        if not self.totalDict :
            self.wordsFrequency()
        sortedDictList = [(key, val) for key, val in self.totalDict.items()]
        sortedDictList.sort(key = lambda e : e[1], reverse = not reverse)
        self.sortedList = sortedDictList
    
    def getWordsFrequency(self, reverse = False) :
        self.getSortedWordsFrequency(reverse)
        print(self.sortedList)
        
if __name__ == "__main__" :
    w4 = SearchEngineWithOrderedWebWords('http://www.times.com', 
                                         'https://www.amazon.com', 'https://github.com')
    w4.getSortedWordsFrequency()
    for i in w4 :
        print(i)
```
