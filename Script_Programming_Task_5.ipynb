  - ÏöîÍµ¨ÏÇ¨Ìï≠ 1. ÏÉùÏÑ±ÏûêÏóê count Í∞íÍ≥º step Í∞íÏùÑ Ïù∏ÏûêÎ°ú Î∞õÏùÑ Ïàò ÏûàÎã§.

  - ÏöîÍµ¨ÏÇ¨Ìï≠ 2. Îã§ÏùåÍ≥º Í∞ôÏù¥ CounterÏùò Ïù∏Ïä§ÌÑ¥Ïä§Î•º Ï∂úÎ†•ÏùÑ Ìï¥Ï£ºÎäî __str__() Î©îÏÜåÎìúÎ•º Counter ÌÅ¥ÎûòÏä§ ÎÇ¥Ïóê Íµ¨ÌòÑÌïòÏãúÏò§.

  - ÏöîÍµ¨ÏÇ¨Ìï≠ 6. Îã§ÏùåÍ≥º Í∞ôÏùÄ Í¥ÄÍ≥Ñ Ïó∞ÏÇ∞ (>, <, ==)Ïù¥ ÏàòÌñâÎê† Ïàò ÏûàÎèÑÎ°ù Counter ÌÅ¥ÎûòÏä§ ÎÇ¥Ïóê Í¥ÄÎ†® Î©îÏÜåÎìúÎ•º Ï∂îÍ∞ÄÌïòÏãúÏò§.
    > \>>> print(c) > 10

    \>>> print(d) < 10

    True  
    \>>> print(d) != 9

```python
class Counter() :
    # ÏöîÍµ¨ÏÇ¨Ìï≠ 1
    def __init__(self, count, step = 1) :
        self.count = count
        self.step = step
        
    # ÏöîÍµ¨ÏÇ¨Ìï≠ 2
    def __str__(self) :
        return '[Count (step: {})] {}'.format(self.step, self.count)
    
    # ÏöîÍµ¨ÏÇ¨Ìï≠ 3
    def incr(self) :
        self.count += self.step
        
    # ÏöîÍ∂áÏÇ¨Ìï≠ 4
    def __call__(self) :
        self.incr()
        
    # ÏöîÍµ¨ÏÇ¨Ìï≠ 5
    def __add__(self, other) :
        if type(other) == Counter :
            return Counter(self.count + other.count, self.step)
        else :
            return Counter(self.count + other, self.step)
        
    def __sub__(self, other) :
        if type(other) == Counter :
            return Counter(self.count - other.count, self.step)
        else :
            return Counter(self.count - other, self.step)
        
    # ÏöîÍµ¨ÏÇ¨Ìï≠ 6
    def __lt__(self, other) :
        if type(other) == Counter :
            return self.count < other.count
        else :
            return self.count < other
        
    def __gt__(self, other) :
        if type(other) == Counter :
            return self.count > other.count
        else :
            return self.count > other
        
    def __eq__(self, other) :
        if type(other) == Counter :
            return self.count == other.count and self.step == other.step
        else :
            return self.count == other
        
c = Counter(10)
d = Counter(10, 2)

print(c); print(d)
print()

c.incr(); d.incr()
print(c); print(d)
print()

c(); d()
print(c); print(d)
print()

c = c + 5; d = d - 5
print(c); print(d)
print()

print(c > 10); print(d > 10)
print(c < 10); print(d < 10)
print(c == 17); print(d != 9)
```

4. Îã§ÏùåÏùÄ ÎÇ¥Ïû• ÏûêÎ£åÌòï listÎ•º ÏÑúÎ∏åÌÅ¥ÎûòÏã±ÌïòÏó¨ ÎßåÎì† MySet ÌÅ¥ÎûòÏä§ Ï†ïÏùò ÎÇ¥Ïö©Ïù¥Îã§. Îã§Ïùå ÌÅ¥ÎûòÏä§ Ï†ïÏùòÏóêÏÑú __init__(), __str__(), eliminate_duplicate()Ïùò ÏÑ∏ Í∞úÏùò Î©îÏÜåÎìú ÏΩîÎìú ÎÇ¥Ïö©ÏùÑ ÏûêÏã†Ïù¥ Îã§Î•∏ ÏÇ¨ÎûåÏóêÍ≤å Í∞ÄÎ•¥ÏπúÎã§Í≥† ÏÉùÍ∞ÅÌïòÎ©∞ ÏÑ§Î™ÖÌï¥Î≥¥ÏãúÏò§.
  - MySetÏùÄ ÏßëÌï©(Set) ÏûêÎ£åÌòïÏùÑ Ï†ïÏùòÌïòÎ†§Îäî ÏùòÎèÑÌïòÏóê ÎßåÎì§Ïñ¥ÏßÑ ÌÅ¥ÎûòÏä§Ïù¥Îã§.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
            
    def __or__(self, other) :
        if type(other) == MySet :
            return MySet(self + other)
            
    def __and__(self, other) :
        if type(other) == MySet :
            newList = [e for e in self if e in other]
            return MySet(newList)
        
    def __sub__(self, other) :
        if type(other) == MySet :
            andList = self & other
            subList = [e for e in self if e not in andList]
            return MySet(subList)
        
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    u = s | t
    print(u)
    u = s & t
    print(u)
    s = MySet([1, 2, 3])
    t = MySet([3, 4, 5])
    u = s - t
    print(u)
```

```python
import requests
import string
import itertools
import collections

class SearchEngine :
    def __init__(self, *urls) :
        self.urls = [e for e in urls]
        self.dictList = []
        self.totalDict = dict()
        
    def addUrl(self, url) :
        self.urls.append(url)
        
    def removeUrl(self, url) :
        if url in self.urls : self.urls.remove(url)
        else : return
        
    def listUrls(self) :
        for e in self.urls : print(e)
    
    def wordsFrequency(self) :
        for each in self.urls :
            req = requests.get(each)
            source = req.text
            punct = string.punctuation
            
            htmlTag = cssJvsc = False
            pureStr = ''
            pureStrList = []
            for i in range(len(source)) :
                css = jvsc = ''
                if source[i] == '<' and not htmlTag :
                    pureStr += ' '
                    if i + 8 <= len(source) :
                        for j in range(i + 1, i + 7) : css += source[j]
                        for k in range(i + 1, i + 8) : jvsc += source[k]
                    if css[ : -1] == 'style' or jvsc[ : -1] == 'script' :
                        cssJvsc = True
                    elif css == '/style' or jvsc == '/script' and cssJvsc :
                        cssJvsc = False
                    if not cssJvsc : htmlTag = True
                elif source[i] == '>' and htmlTag :
                    htmlTag = False
                    continue
                if not htmlTag and not cssJvsc :
                    pureStr += source[i]
                    
            pureStrListS = pureStr.split()
            for i in range(len(pureStrListS)) :
                pureStrListS[i] = pureStrListS[i].strip().strip(punct)
                
            for i in range(len(pureStrListS)) :
                for j in punct :
                    while j in pureStrListS[i] :
                        for k in range(len(pureStrListS[i])) :
                            indx = pureStrListS[i].find(j)
                            pureStrListS[i] = pureStrListS[i][ : indx] + ' ' + \
                            pureStrListS[i][indx + 1 : ]
                            break
            pureStrList = ' '.join(pureStrListS).split()
            
            blkCount = pureStrList.count('')
            for i in range(blkCount) :
                pureStrList.remove('')
            
            wordDict = {word : 0 for word in pureStrList}
            for word in wordDict.keys() :
                for val in pureStrList :
                    if word == val : wordDict[word] += 1
                        
            with open('stop_words.txt', 'r', encoding = 'utf-8') as f :
                stopWordList = f.readlines()
                for i in range(len(stopWordList)) :
                    stopWordList[i] = stopWordList[i][ : -1]
                deleteWordList = [keys for word in stopWordList
                                 for keys in wordDict 
                                 if word == keys or word == keys.lower()]
                
                tempSet = {element for element in deleteWordList}
                deleteWordList = [element for element in tempSet]
                for word in deleteWordList :
                    del wordDict[word]
                    
            self.dictList.append(wordDict)

        sumValList = collections.defaultdict(list)
        for i in range(len(self.dictList)) :
            for key, val in itertools.chain(self.dictList[i].items()) :
                sumValList[key].append(val)

        self.totalDict = {key : 0 for key in sumValList.keys()}
        for key, val in sumValList.items() :
            for element in range(len(val)) :
                self.totalDict[key] += val[element]

        #print(len(self.totalDict))
        #print(self.totalDict)
        return self.totalDict
        
    def getWordsFrequency(self) :
        if not self.totalDict :
            self.wordsFrequency()
        print(self.totalDict)
    
    def getMaxFrequencyWord(self) :
        if not self.totalDict :
            self.wordsFrequency()
        sortedDict = [(key, val) for key, val in self.totalDict.items()]
        sortedDict.sort(key = lambda e : e[1], reverse = True)
        maxFrequencyDict = {element[0] : element[1] for element in sortedDict
                           if element[1] == sortedDict[0][1]}
        print(maxFrequencyDict)
    
    def searchUrlByWord(self, word) :
        wordSplitList = word.split()
        with open('stop_words.txt', 'r', encoding = 'utf-8') as f :
            stopWordList = f.readlines()
            for i in range(len(stopWordList)) :
                stopWordList[i] = stopWordList[i][ : -1]
            deleteWordList = [words for sWord in stopWordList
                              for words in wordSplitList if sWord == words or sWord == words.lower()]
            
            tempSet = {element for element in deleteWordList}
            deleteWordList = [element for element in tempSet]
            for words in deleteWordList :
                wordSplitList.remove(words)
                
        similarityList = []
        for i in range(len(self.urls)) :
            for element in wordSplitList :
                if element in self.dictList[i] :
                    similarityList.append((self.urls[i], self.dictList[i][element]))
                elif element.lower() in self.dictList[i] :
                    similarityList.append((self.urls[i], self.dictList[i][element.lower()]))
                elif element.upper() in self.dictList[i] :
                    similarityList.append((self.urls[i], self.dictList[i][element.upper()]))
                elif element.title() in self.dictList[i] :
                    similarityList.append((self.urls[i], self.dictList[i][element.title()]))
                else : similarityList.append((self.urls[i], 0))
                    
        #print(similarityList)
        similarityList.sort(key = lambda e : e[1], reverse = True)
        #print(similarityList)
        print(word)
        for links in similarityList :
            if similarityList[0][1] == 0 :
                print("Í≤ÄÏÉâ Í≤∞Í≥ºÍ∞Ä ÏóÜÏäµÎãàÎã§.")
                break
            if links[1] == similarityList[0][1] :
                print(links[0])
            
                
if __name__ == '__main__' :
    w1 = SearchEngine('http://www.cnn.com', 'http://www.times.com', 'https://www.amazon.com')
    w2 = SearchEngine('http://www.cnn.com', 'http://www.times.com')
    w3 = SearchEngine()
    w1.addUrl('https://github.com')
    w3.addUrl('http://stackoverflow.com')
    w1.removeUrl('http://www.cnn.com')
    w2.removeUrl('http://stackoverflow.com')
    print('-' * 80)
    print("w1's url list")
    w1.listUrls()
    print('-' * 80)
    print("w2's url list")
    w2.listUrls()
    print('-' * 80)
    print("w3's url list")
    w3.listUrls()
    print('-' * 80)
#     w1.getWordsFrequency()
#     w2.getWordsFrequency()
#     w3.getWordsFrequency()
    w1.getMaxFrequencyWord()
    print('-' * 80)
    w2.getMaxFrequencyWord()
    print('-' * 80)
    w3.getMaxFrequencyWord()
    print('-' * 80)
    w1.searchUrlByWord("New York")
    print()
    w1.searchUrlByWord("news")
    print()
    w1.searchUrlByWord("github")
    print()
    w1.searchUrlByWord("studio")
    print('-' * 80)
    w2.searchUrlByWord("CNN")
    print()
    w2.searchUrlByWord("times")
    print('-' * 80)
    w3.searchUrlByWord("Stack")
    print()
    w3.searchUrlByWord("Python")
    print('-' * 80)
```
